```python
import pandas as pd
import calendar
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Average Step Count by Month

```python
stepCountsDB = pd.read_csv("HKQuantityTypeIdentifierStepCount_2023-11-329_14-06-23_SimpleHealthExportCSV.csv",
                           header=1,usecols=[5,6,7,8])

stepCountsDB['date'] = pd.to_datetime(stepCountsDB['startDate'])

stepCountsDB_GroupedByMonth = stepCountsDB.groupby(stepCountsDB['date'].dt.to_period("M"))['value'].sum().reset_index()

stepCountsDB_GroupedByMonth['avg'] = stepCountsDB_GroupedByMonth.apply(
    lambda row: row['value'] / calendar.monthrange(row['date'].year, row['date'].month)[1],
    axis=1
)
stepCountsDB_GroupedByMonth['avg'] = stepCountsDB_GroupedByMonth['avg'].astype(int)

stepCountsDB_GroupedByMonth = stepCountsDB_GroupedByMonth[stepCountsDB_GroupedByMonth['date'].between('2021-10', '2023-11', inclusive='both')].reset_index()


dateList = []

for i in stepCountsDB_GroupedByMonth["date"]:
    dateList.append(str(i))

plt.figure(figsize=(14, 6))
plt.xticks(rotation=45)
# Bar chart for the average divided by days
plt.plot(dateList, stepCountsDB_GroupedByMonth['avg'],marker ="o")
plt.title('Average Step Count by Months')
plt.xlabel('Month')
plt.ylabel('Step Count')

plt.tight_layout()
plt.show()
```
